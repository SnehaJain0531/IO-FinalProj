```python
import csv
from bs4 import BeautifulSoup
import lxml
import urllib.request
import pandas
```

```python
html=urllib.request.urlopen('https://karki23.github.io/Weather-Data/assignment.html').read()
```

```python
obj=BeautifulSoup(html,'lxml')
```

```python
web=[]
links=obj.find_all('a')
for i in links:
    web.append(i.get('href'))
print(web)
```

```python
counter=1
for k in web:
    h=urllib.request.urlopen('https://karki23.github.io/Weather-Data/'+ k).read()
    a=BeautifulSoup(h,'lxml')
    rows=[]
    t=a.find_all('table')
    for i in t:
        tr=i.find_all('tr')
        th=[j.text for j in tr[0].find_all('th')]
        rows.append(th)
        for m in tr[1:]:
            td=[l.text for l in m.find_all('td')]
            rows.append(td)
    print(counter)
    with open('dataset.csv','a') as f:
        writer=csv.writer(f)
        for i in rows:
            writer.writerow(i)
    counter=counter+1
```
